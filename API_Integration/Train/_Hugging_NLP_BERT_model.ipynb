Paso 2. Tokenizar
Descargando un tokenizadore desde el repositorio ( Exploratorio)

```python
from transformers import AutoTokenizer
repo_id = "bert-base-cased"
tokenizer = AutoTokenizer.from_pretrained(repo_id)
```

**BERT: 'input_ids'**



*   101: Inicio de una frase
*   102: Final de la secuencia  la frase
* 117: es ","
* 119: es "."
* 131: es ":"
* Los demás códigos corresponden a las palabras tokenizadas


**BERT: 'token_type_ids'** Determinan si el texto corresponde a una misma frase



*   0: Es la misma frase
*   1: No es la misma o secuerncia


**BERT: 'attention_mask'** Si los tokens son originales de la palabra o son rellenos



*   0: si el token es un relleno (padding)
*   1: si es token propio de la secuencia


```python
token_frase_1= tokenizer(data["train"]["sentence1"][200])
token_frase_1

##Enconder
# pasamos de texto a numeros una codificación
```

```python
#Decoder
tokenizer.convert_ids_to_tokens(token_frase_1["input_ids"])
```

Paso 3. Preprocesamiento del data set con el tokenizador BERT

Se genera el data set preprocesado

```python
# funcion para tokenizar las frases

def token_sentences(frases):
  return tokenizer(frases["sentence1"], frases["sentence2"], truncation=True)
```

```python
# el .map aplica la funcion a toda la data original yt regresada el nuevo data set
data_prepro = data.map(token_sentences, batched=True)
```

```python
print(data_prepro["train"][57])
print(data["train"][57])
```

Paso 3.1. Contiunuamos el preprocesamiento con padding ( relleno) de las frases

Asegurando que todas las frases tengan la mismaa longitud

```python
from transformers import DataCollatorWithPadding
```

```python
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

Paso 4. La estructura para entrenamiento y evaluación

```python
import numpy as np
import evaluate
```

```python
def compute_metrics(eval_pred):
  metric = evaluate.load("glue", "mrpc")
  logits, labels = eval_pred
  predictions = np.argmax(logits, axis=-1)
  return metric.compute(predictions=predictions, references=labels)
```

Paso 5. Configurando el Train

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
```

```python
labels = data["train"].features["label"].names
```

```python
model = AutoModelForSequenceClassification.from_pretrained(repo_id,
                                                           num_labels=len(labels), # Número de clases en este caso son 2
                                                           id2label={str(i): c for i, c in enumerate(labels)},  # "0" : "NO-EQUIVALENT"
                                                           label2id={c: str(i) for i, c in enumerate(labels)}) # "NO-EQUIVALENT" : "0"
```

Paso 6. Configuración de Hugging face para entrenamiento

```python
from transformers import TrainingArguments

trainingArg = TrainingArguments(
    output_dir= "./NLP_model_Deisy_BIT_2",
    evaluation_strategy= "steps",
    num_train_epochs=4,
    push_to_hub = True,
    load_best_model_at_end= True
)
```

```python
!huggingface-cli login
```

Paso 7. Estructura del Train

```python
trainer = Trainer(
    model=model,
    args=trainingArg,
    train_dataset=data_prepro["train"],
    eval_dataset = data_prepro["validation"],
    data_collator= data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)
```

Paso 8. Train

```python
train_results = trainer.train()
trainer.save_model()
trainer.log_metrics("train", train_results.metrics)
trainer.save_metrics("train", train_results.metrics)
trainer.save_state()
```
